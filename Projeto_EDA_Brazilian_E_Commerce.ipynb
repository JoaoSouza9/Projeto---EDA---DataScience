###Criar uma coluna categórica com valores como “pequeno”, “médio”, e “grande” para cada produto, a depender de seus valores de height e width.


Para enriquecer ainda mais a classificação, e tomando proveito do fato de que no Dataset existe o campo 'product_length_cm', a classificação será feita tomando por base o volume do produto.

Da análise feita, temos que dois produtos possuem campos 'product_length_cm', 'product_width_cm' e 'product_height_cm' como "Não Números", ou seja, NaN. E por isso não foi possível classificá-los de acorso com essa categorização.



```python
productsdf['product_volume_cm'] = productsdf['product_length_cm'] * productsdf['product_height_cm'] * productsdf['product_width_cm']
productsdf
```

```python
condicoes = [(productsdf['product_volume_cm'] < 2880),
             ((productsdf['product_volume_cm'] >= 2880) & (productsdf['product_volume_cm'] <= 18480)),
             (productsdf['product_volume_cm'] > 18480)]
opcoes = ['pequeno', 'medio', 'grande']
productsdf['volume_classification'] = np.select(condicoes, opcoes)
productsdf
```

```python
productsdf['volume_classification'].value_counts()
```

```python
find_values = productsdf[productsdf['volume_classification'] == '0']
find_values
```
